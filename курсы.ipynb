```python
health = 100

def take_damage(damage):
    global health
    health -= damage
    print(f"Нанесено {damage} урона")

def heal(amount):
    global health
    health += amount
    print(f"Получено {amount} здоровья")

def is_alive():
    return health > 0

while True:
    print("\n=== Герой ===")
    print("1 — Получить урон")
    print("2 — Получить лечение")
    print("3 — Проверить жизнь")
    print("4 — Посмотреть здоровье")
    print("5 — «Битва с драконом»")
    print("6 — Выход")

    choice = input("Выберите действие (1–6): ")

    if choice == "1":
        damage = int(input("Введите значение урона: "))
        take_damage(damage)
    
    elif choice == "2":
        amount = int(input("Введите количество здоровья для восстановления: "))
        heal(amount)

    elif choice == "3":
        alive = is_alive
        if alive:
            print("Герой жив?")
        else:
            print("Герой погиб.")

    elif choice == "4":
        print("Текущее здоровье:", health)

    elif choice == "5":
        print("\n🔥 Битва с драконом!")
        take_damage(30)
        heal(10)
        alive = is_alive()
        print("Выживший герой:" if alive else "Ваш герой погиб...")

    elif choice == "6":
        print("До свидания!")
        break
    else:
        print("Неверный выбор. Введите число от 1 до 6.")
```

```python
email = input("Введите email: ")

if " " in email:
    print("Email некорректный: содержит пробелы")
elif "@" not in email:
    print("Email некорректный: отсутствует символ @")
else:
    at_position = email.find("@")
    if "." in email[at_position:]:
        print("Email корректный")
    else:
        print("Email некорректный: отсутствует точка после @")
```

```python
def validate_email(email):

    # Убираем лишние пробелы по краям, если пользователь случайно их ввел
    email = email.strip()

    # 1. Проверка на пробелы внутри строки
    if " " in email:
        return False, "Email не должен содержать пробелы."

    # 2. Проверка наличия и количества символов '@'
    if email.count("@") != 1:
        return False, "Email должен содержать ровно один символ '@'."

    # Разделяем адрес на имя пользователя и доменную часть
    user_part, domain_part = email.split("@")

    # 3. Проверка, что имя пользователя не пустое
    if not user_part:
        return False, "Отсутствует имя пользователя перед '@'."

    # 4. Проверка наличия точки в доменной части и что она не последняя
    if "." not in domain_part:
        return False, "Доменная часть должна содержать точку (например, .com)."
    
    if domain_part.startswith(".") or domain_part.endswith("."):
        return False, "Точка не может быть первым или последним символом домена."

    return True, "Email корректный"

def main():
    print("=== Программа проверки Email ===")
    
    while True:
        try:
            user_input = input("\nВведите email для проверки: ")
            
            is_valid, message = validate_email(user_input)
            
            if is_valid:
                print(f"✅ Результат: {message}")
            else:
                print(f"❌ Ошибка: {message}")
            
            # Предлагаем повторить ввод
            retry = input("\nХотите проверить еще раз? (Да/Нет): ").strip().lower()
            if retry not in ["да", "yes", "y", "д"]:
                print("Программа завершена. До свидания!")
                break
                
        except Exception as e:
            print(f"Произошла непредвиденная ошибка: {e}")

if __name__ == "__main__":
    main()
```

```python
import json

# Шаг 2. Функция загрузки данных
def load_heroes():
    try:
        with open("heroes.json", "r", encoding='utf-8') as file:
            return json.load(file)
    except FileNotFoundError:
        print("Файл не найден. Убедитесь, что файл heroes.json существует.")
        return []
    except json.JSONDecodeError:
        print("Ошибка чтения файла. Файл поврежден или имеет неверный формат.")
        return []

# Шаг 3. Функция сохранения данных
def save_heroes(heroes):
    with open("heroes.json", "w", encoding='utf-8') as file:
        json.dump(heroes, file, ensure_ascii=False, indent=4)

# Шаг 4. Функция вывода героев
def show_heroes(heroes):
    print("\n--- Герои ---")
    if not heroes:
        print("Нет героев в списке.")
    else:
        for hero in heroes:
            print(f"{hero['name']}, уровень: {hero['level']}, сила: {hero['power']}")

# Шаг 5. Функция добавления героя
def add_hero():
    name = input("Введите имя героя: ").strip()
    
    # Проверка ввода уровня
    while True:
        try:
            level = int(input("Введите уровень героя (целое число): "))
            break
        except ValueError:
            print("Ошибка! Введите целое число.")
    
    # Проверка ввода силы
    while True:
        try:
            power = float(input("Введите силу героя (дробное число): "))
            break
        except ValueError:
            print("Ошибка! Введите число (можно дробное).")
    
    # Создание словаря с героем
    new_hero = {
        "name": name,
        "level": level,
        "power": power
    }
    
    # Загрузка текущего списка и добавление нового героя
    heroes = load_heroes()
    heroes.append(new_hero)
    
    # Сохранение обновленного списка
    save_heroes(heroes)
    print("Герой добавлен!")

# Шаг 6. Функция удаления героя
def remove_hero():
    name_to_remove = input("Какого героя удалить? ").strip()
    
    heroes = load_heroes()
    updated_heroes = [h for h in heroes if h["name"] != name_to_remove]
    
    if len(updated_heroes) < len(heroes):
        save_heroes(updated_heroes)
        print("Герой удален!")
    else:
        print("Ошибка: такого героя нет.")

# Шаг 10. Функция битвы с драконом
def battle_with_dragon(heroes):
    name = input("Какой герой будет сражаться с драконом? ").strip()
    
    found = False
    for hero in heroes:
        if hero["name"].lower() == name.lower():
            found = True
            if hero["power"] > 70:
                print(f"{hero['name']} победил дракона!")
            else:
                print(f"{hero['name']} проиграл. Дракон слишком силён.")
    
    if not found:
        print("Такого героя нет.")

# Шаг 7. Главное меню программы
def main_menu():
    while True:
        print("\n=== База героев фэнтези ===")
        print("1 — Посмотреть всех героев")
        print("2 — Добавить героя")
        print("3 — Удалить героя")
        print("4 — Сразиться с драконом")
        print("5 — Выход")
        
        choice = input("Выберите действие: ").strip()
        
        if choice == "1":
            heroes = load_heroes()
            show_heroes(heroes)
        elif choice == "2":
            add_hero()
        elif choice == "3":
            remove_hero()
        elif choice == "4":
            heroes = load_heroes()
            battle_with_dragon(heroes)
        elif choice == "5":
            print("До встречи! Все изменения сохранены.")
            break
        else:
            print("Неверный выбор. Попробуйте снова.")

# Шаг 8. Запуск программы
if __name__ == "__main__":
    main_menu()
```

```python
import random
from datetime import datetime
from collections import Counter
import os

# Инициализация
stats = Counter()
filename = "dice_rolls.txt"

# Загрузка истории из файла
if os.path.exists(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            for line in file:
                if '=' in line:
                    try:
                        sum_value = int(line.split('=')[1].strip())
                        stats[sum_value] += 1
                    except (ValueError, IndexError):
                        continue
    except Exception as e:
        print(f"Ошибка при загрузке истории: {e}")

print("=== Имитация броска двух шестигранных костей ===")

while True:
    print("\nВыберите команду:")
    print("1. Бросить кости")
    print("2. Показать статистику")
    print("3. Выход")
    
    choice = input("> ").strip()
    
    if choice == '1':
        # Бросок костей
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        total = die1 + die2
        
        print(f"\nВы бросили кости: {die1} + {die2} = {total}")
        
        # Сохранение в файл
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        roll_record = f"[{timestamp}] Бросок: {die1} + {die2} = {total}\n"
        
        try:
            with open(filename, 'a', encoding='utf-8') as file:
                file.write(roll_record)
            stats[total] += 1
            print("Результат записан в файл.")
        except Exception as e:
            print(f"Ошибка при сохранении в файл: {e}")
    
    elif choice == '2':
        # Показ статистики
        if not stats:
            print("\nСтатистика бросков:")
            print("Пока нет ни одного броска.")
        else:
            print("\nСтатистика бросков:")
            # Сортируем суммы по возрастанию
            for total in sorted(stats.keys()):
                count = stats[total]
                # Склоняем слово "раз" в зависимости от числа
                if count % 10 == 1 and count % 100 != 11:
                    times_word = "раз"
                else:
                    times_word = "раз"
                print(f"{total} — {count} {times_word}")
            
            # Показываем общее количество бросков
            total_rolls = sum(stats.values())
            print(f"\nВсего бросков: {total_rolls}")
    
    elif choice == '3':
        print("\nДо свидания!")
        break
    
    else:
        print("\nНеверная команда. Пожалуйста, выберите 1, 2 или 3.")
```

```python
import random
import time
from datetime import datetime
import os

print("🎮 Угадай число с таймером!")
print("Правила:")
print("1. Я загадал число от 1 до 100.")
print("2. У тебя есть 60 секунд.")
print("3. Постарайся угадать как можно быстрее.\n")

secret_number = random.randint(1, 100)
attempts = 0
game_won = False 
guess = None
start_time = datetime.now()
print(f"⏰ Начало игры: {start_time.strftime('%H:%M:%S')}\n")

print("Начинаем! У вас 60 секунд.")

while True:
    try:
        user_input = input("Введите число: ")
        guess = int(user_input)
        attempts += 1

        if guess < secret_number:
            print("Маловато...")
        elif guess > secret_number:
            print("Перебор!")
        else:
            print("🎉 Вы угадали!")
            game_won = True
            break

        elapsed = (datetime.now() - start_time).seconds
        if elapsed >= 60:
            print(f"⏳ Время вышло! Загаданное число было: {secret_number}")
            break
            
    except ValueError:
        print("Ошибка: введите целое число.")

end_time = datetime.now()
elapsed_seconds = (end_time - start_time).seconds

print(f"\n⏰ Конец игры: {end_time.strftime('%H:%M:%S')}")
print(f"⏱️  Вы потратили: {elapsed_seconds} секунд")
print(f"🔢 Число попыток: {attempts}")

with open("game_result.txt", "w", encoding='utf-8') as file:
    file.write("🎮 Результат игры 'Угадай число'\n")
    file.write(f"Дата: {end_time.strftime('%Y-%m-%d %H:%M')}\n")
    file.write(f"Загаданное число: {secret_number}\n")
    file.write(f"Число попыток: {attempts}\n")
    file.write(f"Затраченное время: {elapsed_seconds} секунд\n")

    if game_won:
        status = "Победа"
    else:
        status = "Поражение"
    file.write(f"Статус: {status}")

print("\n✅ Результат сохранён в файл game_result.txt")
```

```python
numbers = []
working_list = []

while True:
    print("\nВыберите команду:")
    print("1. Ввести список чисел")
    print("2. Сортировать по возрастанию")
    print("3. Сортировать по убыванию")
    print("4. Показать список")
    print("5. Выход")
    
    choice = input("> ")

    if choice == '1':
        try:
            numbers = list(map(int, input("Введите числа через пробел: ").split()))
            working_list = numbers[:]
            print(f"Список сохранён: {numbers}")
        except ValueError:
            print("Ошибка: вводите только целые числа.")

    elif choice == '2':
        if numbers:
            working_list.sort()
            print(f"Список отсортирован по возрастанию: {working_list}")
        else:
            print("Список пуст!")

    elif choice == '3':
        if numbers:
            working_list = sorted(working_list, reverse=True)
            print(f"Список отсортирован по убыванию: {working_list}")
        else:
            print("Список пуст!")

    elif choice == '4':
        if numbers:
            print(f"Исходный список: {numbers}")
            print(f"Отсортированный список: {working_list}")
        else:
            print("Списки пусты.")

    elif choice == '5':
        print("Выход из программы.")
        break
    else:
        print("Неверный ввод, попробуйте снова.")
```

```python
import json

# Попытка прочитать существующий файл
try:
    with open('books.json', 'r', encoding='utf-8') as f:
        books = json.load(f)
    print("Файл books.json найден. Текущие книги:")
    for b in books:
        print(f"  {b['title']} - {b['year']}")

    # Спрашиваем, хочет ли пользователь добавить книги
    choice = input("Хотите добавить новые книги? (да/нет): ").strip().lower()
    if choice == 'да':
        new_books = []
        while True:
            title = input("Введите название книги (или 'стоп' для завершения): ").strip()
            if title.lower() == 'стоп':
                break
            try:
                year = int(input("Введите год издания: "))
            except ValueError:
                print("Ошибка: год должен быть числом. Попробуйте снова.")
                continue
            new_books.append({"title": title, "year": year})

        books.extend(new_books)
        # Сохраняем обновлённый список
        with open('books.json', 'w', encoding='utf-8') as f:
            json.dump(books, f, ensure_ascii=False, indent=2)
        print("Новые книги добавлены и файл обновлён.")
    else:
        print("Работаем с существующим списком.")

except FileNotFoundError:
    print("Файл books.json не найден. Создадим новый список книг.")
    books = []
    while True:
        title = input("Введите название книги (или 'стоп' для завершения): ").strip()
        if title.lower() == 'стоп':
            break
        try:
            year = int(input("Введите год издания: "))
        except ValueError:
            print("Ошибка: год должен быть числом. Попробуйте снова.")
            continue
        books.append({"title": title, "year": year})

    # Сохраняем в файл
    with open('books.json', 'w', encoding='utf-8') as f:
        json.dump(books, f, ensure_ascii=False, indent=2)
    print("Список книг сохранён в файл books.json.")

# Сортируем книги по году издания
books.sort(key=lambda book: book['year'])

# Выводим результат
print("\nОтсортированный список книг по году издания:")
for book in books:
    print(f"  {book['title']} — {book['year']}")
```

```python
import requests

def get_cat_fact():
    response = requests.get("https://catfact.ninja/fact")
    if response.status_code == 200:
        data = response.json()
        print("😺", data["fact"])
    else:
        print("Ошибка загрузки факта.")

def get_random_joke():
    response = requests.get("https://official-joke-api.appspot.com/random_joke")
    if response.status_code == 200:
        joke = response.json()
        print("😄", joke["setup"])
        print("😄", joke["punchline"])
    else:
        print("Ошибка загрузки шутки.")

def get_random_user():
    response = requests.get("https://randomuser.me/api/")
    if response.status_code == 200:
        data = response.json()

        if data.get('results') and len(data['results']) > 0:
            user = data['results'][0]
            gender = user['gender']
            name_info = user['name']
            full_name = f"{name_info['title']} {name_info['first']} {name_info['last']}"
            email = user['email']
            location = f"{user['location']['city']}, {user['location']['country']}"

            print(f"🧍 Пол: {gender}")
            print(f"🧍 Имя: {full_name}")
            print(f"🧍 Email: {email}")
            print(f"🧍 Местоположение: {location}")
        else:
            print("Ошибка: не удалось получить данные пользователя")
    else:
        print("Ошибка загрузки пользователя.")

def get_dog_image():
    response = requests.get("https://dog.ceo/api/breeds/image/random")
    if response.status_code == 200:
        message = response.json()
        print("🐶", message["message"])
    else:
        print("Ошибка загрузки картинки.")

while True:
    print("\n=== Мир случайностей ===")
    print("1 — Факт о котике")
    print("2 — Случайная шутка")
    print("3 — Информация о случайном пользователе")
    print("4 — Случайное фото собаки")
    print("5 — Выход")

    choice = input("Выберите действие (1–5): ")

    if choice == "1":
        get_cat_fact()
    elif choice == "2":
        get_random_joke()
    elif choice == "3":
        get_random_user()
    elif choice == "4":
        get_dog_image()
    elif choice == "5":
        print("До встречи!")
        break
    else:
        print("Неверный выбор. Попробуйте снова.")
```

```python
import requests

def get_user_info(user_id):
    """
    Получает информацию о пользователе по ID из JSONPlaceholder API
    """
    # URL для запроса
    url = f"https://jsonplaceholder.typicode.com/users/{user_id}"
    
    try:
        # Отправляем GET-запрос
        response = requests.get(url)
        
        # Проверяем статус ответа
        if response.status_code == 200:
            # Преобразуем ответ в JSON
            user_data = response.json()
            
            # Выводим информацию о пользователе
            print(f"Имя: {user_data['name']}")
            print(f"Email: {user_data['email']}")
            print(f"Город: {user_data['address']['city']}")
            
        elif response.status_code == 404:
            print(f"Пользователь с ID {user_id} не найден.")
        else:
            print(f"Ошибка при получении данных. Код статуса: {response.status_code}")
            
    except requests.exceptions.RequestException as e:
        print(f"Ошибка соединения: {e}")
    except KeyError as e:
        print(f"Ошибка в структуре данных: отсутствует поле {e}")
    except Exception as e:
        print(f"Произошла непредвиденная ошибка: {e}")

def main():
    """
    Основная функция программы
    """
    try:
        # Запрашиваем ID пользователя
        user_id = input("Введите ID пользователя: ").strip()
        
        # Проверяем, что введено число
        if not user_id.isdigit():
            print("Ошибка: ID должен быть числом.")
            return
        
        # Преобразуем в целое число
        user_id = int(user_id)
        
        # Получаем информацию о пользователе
        get_user_info(user_id)
        
    except KeyboardInterrupt:
        print("\nПрограмма прервана пользователем.")
    except Exception as e:
        print(f"Произошла ошибка: {e}")

if __name__ == "__main__":
    main()
```

```python
notes = [1]

def load_notes():
    """Загружает заметки из файла"""
    try:
        with open("notes.txt", "r") as file:
            return file.read().splitlines()
    except FileNotFoundError:
        return []

def save_notes(notes):
    """Сохраняет заметки в файл"""
    with open("notes.txt", "w", encoding="utf-8") as file:
        for note in notes:
            file.write(note + "\n")

def show_notes(notes):
    """Выводит список заметок"""
    if not notes:
        print("Нет заметок.")
    else:
        print("Список заметок:")
        for i, note in enumerate(notes, 1):
            print(f"{i}. {note}")

def add_note(notes):
    """Добавляет новую заметку"""
    note = input("Введите текст заметки: ")
    notes.append(note)
    print("Заметка добавлена!")

def delete_note(notes):
    """Удаляет заметку по номеру"""
    show_notes(notes)
    try:
        index = int(input("Введите номер заметки для удаления: ")) - 1
        if 0 <= index < len(notes):
            removed = notes.pop(index)
            print(f"Заметка '{removed}' удалена")
        else:
            print("Неверный номер заметки")
    except:
        print("ошибка: введите число")

def main():
    """Основной цикл программы"""
    notes = load_notes()

    while True:
        print("\nВыберите команду:")
        print("1 - Посмотреть заметки")
        print("2 - Добавить заметку")
        print("3 - Удалить заметку")
        print("4 - Сохранить и выйти")

        choice = input("Введите номер: ")

        if choice == "1":
            show_notes(notes)
        elif choice == "2":
            add_note(notes)
        elif choice == "3":
            delete_note(notes)
        elif choice == "4":
            save_notes(notes)
            print("Заметки сохранены. До встречи!")
            break
        else:
            print("Неизвестная команда")

if __name__ == "__main__":
    main()
```

```python
import json
import random
import datetime
import os

STATS_FILE = "stats.json"

def load_stats():
    """Загружает статистику из JSON-файла. Если файла нет, возвращает пустой список."""
    if os.path.exists(STATS_FILE):
        try:
            with open(STATS_FILE, "r", encoding="utf-8") as f:
                return json.load(f)
        except (json.JSONDecodeError, IOError):
            print("Ошибка чтения файла статистики. Будет создан новый файл.")
            return []
    return []

def save_stats(stats):
    """Сохраняет статистику в JSON-файл."""
    try:
        with open(STATS_FILE, "w", encoding="utf-8") as f:
            json.dump(stats, f, ensure_ascii=False, indent=4)
    except IOError:
        print("Ошибка сохранения статистики.")

def get_number_input(prompt, min_val, max_val):
    """
    Запрашивает у пользователя целое число в диапазоне [min_val, max_val].
    Повторяет запрос до получения корректного значения.
    """
    while True:
        try:
            value = int(input(prompt))
            if min_val <= value <= max_val:
                return value
            else:
                print(f"Ошибка: введите число от {min_val} до {max_val}.")
        except ValueError:
            print("Ошибка: введите целое число.")

def new_game(stats):
    """
    Проводит одну игру.
    Загадывает число от 1 до 100, даёт 10 попыток.
    После игры добавляет запись в статистику и сохраняет её.
    """
    secret = random.randint(1, 100)
    attempts_left = 10
    print("\n--- Новая игра! ---")
    print("Я загадал число от 1 до 100. У вас есть 10 попыток, чтобы угадать его.")

    while attempts_left > 0:
        print(f"\nОсталось попыток: {attempts_left}")
        guess = get_number_input("Ваше предположение: ", 1, 100)

        if guess == secret:
            print(f"Поздравляю! Вы угадали число {secret} за {10 - attempts_left + 1} попыток!")
            # Запись о победе
            game_result = {
                "date": datetime.datetime.now().isoformat(),
                "secret": secret,
                "attempts_used": 10 - attempts_left + 1,
                "outcome": "win"
            }
            stats.append(game_result)
            save_stats(stats)
            return
        elif guess < secret:
            print("Загаданное число больше.")
        else:
            print("Загаданное число меньше.")

        attempts_left -= 1

    # Если попытки закончились
    print(f"\nК сожалению, вы проиграли. Загаданное число было {secret}.")
    game_result = {
        "date": datetime.datetime.now().isoformat(),
        "secret": secret,
        "attempts_used": 10,
        "outcome": "lose"
    }
    stats.append(game_result)
    save_stats(stats)

def show_stats(stats):
    """Выводит статистику всех сыгранных игр."""
    if not stats:
        print("\nСтатистика пуста. Сыграйте хотя бы одну игру.")
        return

    print("\n--- Статистика игр ---")
    # Определим ширину колонок для красивого вывода
    headers = ["№", "Дата и время", "Загадано", "Попыток", "Результат"]
    rows = []
    for i, game in enumerate(stats, 1):
        date_str = game.get("date", "???")[:16]  # обрежем до минут
        secret = game.get("secret", "?")
        attempts = game.get("attempts_used", "?")
        outcome = "Победа" if game.get("outcome") == "win" else "Поражение"
        rows.append([i, date_str, secret, attempts, outcome])

    # Вывод в виде таблицы
    col_widths = [max(len(str(row[j])) for row in rows + [headers]) for j in range(len(headers))]
    header_line = " | ".join(f"{headers[i]:{col_widths[i]}}" for i in range(len(headers)))
    print(header_line)
    print("-" * len(header_line))
    for row in rows:
        line = " | ".join(f"{str(row[i]):{col_widths[i]}}" for i in range(len(headers)))
        print(line)
    print()

def main():
    """Главное меню приложения."""
    stats = load_stats()
    print("Добро пожаловать в игру 'Угадай число'!")

    while True:
        print("\nМеню:")
        print("1. Новая игра")
        print("2. Показать статистику")
        print("3. Выход")
        choice = input("Выберите действие (1-3): ").strip()

        if choice == "1":
            new_game(stats)
        elif choice == "2":
            show_stats(stats)
        elif choice == "3":
            print("Спасибо за игру! До свидания.")
            break
        else:
            print("Неверный ввод. Пожалуйста, выберите 1, 2 или 3.")

if __name__ == "__main__":
    main()
```

```python
import os

movies = []

def add_movie():
    name = input("Введите название фильма: ").strip()
    if not name:
        print("Название не может быть пустым.")
        return
    
    for movie in movies:
        if movie["name"].lower() == name.lower():
            print("Фильм с таким названием уже есть в списке.")
            return

    genre = input("Введите жанр фильма: ").strip()
    if not genre:
        print("Жанр не может быть пустым.")
        return

    while True:
        try:
            rating = float(input("Введите рейтинг фильма (от 0 до 10): "))
            if 0 <= rating <= 10:
                break
            else:
                print("Рейтинг должен быть в диапазоне от 0 до 10.")
        except ValueError:
            print("Ошибка: введите число (можно дробное, разделитель точка).")

    movie = {"name": name, "genre": genre, "rating": rating}
    movies.append(movie)
    print(f"Фильм '{name}' добавлен.")


def show_all_movies():
    if not movies:
        print("Список фильмов пуст.")
        return

    print("\nСписок всех фильмов:")
    for idx, movie in enumerate(movies, start=1):
        print(f"{idx}. {movie['name']} ({movie['genre']}) - Рейтинг: {movie['rating']}")
    print()


def find_movie():
    """Ищет фильм по точному названию (без учёта регистра)."""
    search = input("Введите название фильма для поиска: ").strip().lower()
    found = False
    for movie in movies:
        if movie["name"].lower() == search:
            print("\nНайден фильм:")
            print(f"  Название: {movie['name']}")
            print(f"  Жанр: {movie['genre']}")
            print(f"  Рейтинг: {movie['rating']}")
            found = True
            break

    if not found:
        print("Фильм не найден.")


def sort_by_rating():
    if not movies:
        print("Список фильмов пуст, нечего сортировать.")
        return

    sorted_movies = sorted(movies, key=lambda x: x["rating"], reverse=True)
    print("\nФильмы, отсортированные по рейтингу (от высокого к низкому):")
    for movie in sorted_movies:
        print(f"{movie['name']} - {movie['rating']}")
    print()


def save_to_file(filename="movies.txt"):
    with open(filename, "w", encoding="utf-8") as f:
        for movie in movies:
            line = f"{movie['name']},{movie['genre']},{movie['rating']}\n"
            f.write(line)
    print(f"Данные сохранены в файл '{filename}'.")


def load_from_file(filename="movies.txt"):
    if not os.path.exists(filename):
        print("Файл с данными не найден. Начинаем с пустого списка.")
        return
    with open(filename, "r", encoding="utf-8") as f:
        for line_num, line in enumerate(f, start=1):
            line = line.strip()
            if not line:
                continue
            parts = line.split(",")
            if len(parts) != 3:
                print(f"Предупреждение: строка {line_num} имеет неверный формат и будет пропущена.")
                continue
            name, genre, rating_str = parts
            try:
                rating = float(rating_str)
            except ValueError:
                print(f"Предупреждение: в строке {line_num} неверное значение рейтинга ('{rating_str}'), строка пропущена.")
                continue
            movies.append({"name": name, "genre": genre, "rating": rating})
        print(f"Данные загружены из файла '{filename}'. Добавлено фильмов: {len(movies)}")

def main():
    load_from_file()

    while True:
        print("\n--- Меню ---")
        print("1. Добавить фильм")
        print("2. Показать все фильмы")
        print("3. Поиск фильма")
        print("4. Отсортировать по рейтингу")
        print("5. Сохранить в файл")
        print("6. Выход")

        choice = input("Выберите действие (1-6): ").strip()

        if choice == "1":
            add_movie()
        elif choice == "2":
            show_all_movies()
        elif choice == "3":
            find_movie()
        elif choice == "4":
            sort_by_rating()
        elif choice == "5":
            save_to_file()
        elif choice == "6":
            print("Выход из программы.")
            break
        else:
            print("Неверный выбор. Пожалуйста, введите число от 1 до 6.")


if __name__ == "__main__":
    main()
```

```python
import json
import os

def load_books():       #загрузка книг из JSON-файла
    if os.path.exists("books.json"):
        with open("books.json", 'r', encoding='utf-8') as file:
            return json.load(file)
    return []

def save_books(books):      #сохранение книг в JSON-файл
    with open("books.json", 'w', encoding='utf-8') as file:
        json.dump(books, file, ensure_ascii=False, indent=2)

def add_book(books):        #добавление новой книги
    title = input("Введите название книги: ")
    author = input("Введите автора: ")
    
    while True:
        try:
            year = int(input("Введите год издания: "))
            break
        except ValueError:
            print("Ошибка! Введите корректный год (число)")
    
    books.append({
        'title': title,
        'author': author,
        'year': year
    })
    save_books(books)
    print(f"Книга '{title}' успешно добавлена!")

def show_all_books(books):      #Показать все книги с сортировкой
    if not books:
        print("Список книг пуст")
        return
    
    print("\nКак отсортировать книги?")
    print("1. По названию")
    print("2. По году")
    
    choice = input("Выберите вариант (1-2): ")
    
    if choice == '1':
        sorted_books = sorted(books, key=lambda x: x['title'])
    elif choice == '2':
        sorted_books = sorted(books, key=lambda x: x['year'])
    else:
        print("Неверный выбор. Показываю без сортировки")
        sorted_books = books
    
    print("\nСписок книг:")
    for book in sorted_books:
        print(f"'{book['title']}' - {book['author']} ({book['year']})")

def search_by_author(books):        #поиск книг по автору
    author = input("Введите имя автора для поиска: ").lower()
    
    found_books = [book for book in books 
                   if author in book['author'].lower()]
    
    if found_books:
        print(f"\nНайдено книг: {len(found_books)}")
        for book in found_books:
            print(f"'{book['title']}' - {book['author']} ({book['year']})")
    else:
        print("Книги этого автора не найдены")

def search_by_year(books):      #поиск книг по году
    try:
        year = int(input("Введите год для поиска: "))
        
        found_books = [book for book in books 
                       if book['year'] == year]
        
        if found_books:
            print(f"\nНайдено книг: {len(found_books)}")
            for book in found_books:
                print(f"'{book['title']}' - {book['author']} ({book['year']})")
        else:
            print("Книги этого года не найдены")
    except ValueError:
        print("Ошибка! Введите корректный год")

def main():
    books = load_books()
    
    while True:
        print("\n=== Управление библиотекой ===")
        print("1. Добавить книгу")
        print("2. Показать все книги")
        print("3. Поиск по автору")
        print("4. Поиск по году")
        print("5. Выход")
        
        choice = input("Выберите действие (1-5): ")
        
        if choice == '1':
            add_book(books)
        elif choice == '2':
            show_all_books(books)
        elif choice == '3':
            search_by_author(books)
        elif choice == '4':
            search_by_year(books)
        elif choice == '5':
            print("До свидания!")
            break
        else:
            print("Неверный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk

def update_display(counters, labels):
    """Обновление отображения всех счётчиков"""
    labels['mouse'].config(text=f"Кликов мышью: {counters[0]}")
    labels['space'].config(text=f"Нажатий пробела: {counters[1]}")
    labels['total'].config(text=f"Всего событий: {counters[2]}")

def on_mouse_click(event, counters, labels):
    """Callback-функция для обработки кликов мыши"""
    counters[0] += 1  # mouse_clicks
    counters[2] += 1  # total_events
    update_display(counters, labels)
    print(f"Клик мышью! Всего кликов: {counters[0]}")
    
    # Визуальный фидбек
    labels['mouse'].config(fg="green")
    event.widget.after(200, lambda: labels['mouse'].config(fg="black"))

def on_space_press(event, counters, labels):
    """Callback-функция для обработки нажатий пробела"""
    counters[1] += 1  # space_presses
    counters[2] += 1  # total_events
    update_display(counters, labels)
    print(f"Нажат пробел! Всего нажатий: {counters[1]}")
    
    # Визуальный фидбек
    labels['space'].config(fg="blue")
    event.widget.after(200, lambda: labels['space'].config(fg="black"))
    
    return "break"

def on_closing(root, counters):
    """Функция корректного закрытия окна"""
    print(f"\n--- ИТОГО ---")
    print(f"Кликов мышью: {counters[0]}")
    print(f"Нажатий пробела: {counters[1]}")
    print(f"Всего событий: {counters[2]}")
    root.destroy()

def main():
    # Создание главного окна
    root = tk.Tk()
    root.title("Счётчик событий")
    root.geometry("300x220")
    
    # Счётчики: [клики_мышью, нажатия_пробела, всего_событий]
    counters = [0, 0, 0]
    
    # Создание меток для отображения данных
    label_mouse = tk.Label(
        root, 
        text=f"Кликов мышью: {counters[0]}", 
        font=("Arial", 12),
        pady=5
    )
    label_mouse.pack()
    
    label_space = tk.Label(
        root, 
        text=f"Нажатий пробела: {counters[1]}", 
        font=("Arial", 12),
        pady=5
    )
    label_space.pack()
    
    label_total = tk.Label(
        root, 
        text=f"Всего событий: {counters[2]}", 
        font=("Arial", 12, "bold"),
        pady=5
    )
    label_total.pack()
    
    # Словарь с метками для удобной передачи
    labels = {
        'mouse': label_mouse,
        'space': label_space,
        'total': label_total
    }
    
    # Привязка событий к callback-функциям с передачей аргументов
    root.bind("<Button-1>", lambda event: on_mouse_click(event, counters, labels))
    root.bind("<space>", lambda event: on_space_press(event, counters, labels))
    
    # Фокус на окно для обработки событий клавиатуры
    root.focus_set()
    
    # Инструкция
    label_instruction = tk.Label(
        root,
        text="Кликните мышью в окне\nили нажмите пробел",
        font=("Arial", 9),
        fg="gray",
        pady=10
    )
    label_instruction.pack()
    
    # Обработка закрытия окна
    root.protocol("WM_DELETE_WINDOW", lambda: on_closing(root, counters))
    
    # Запуск главного цикла приложения
    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk

try:
    with open("record.txt", "r") as file:
        record = int(file.read().strip())
except FileNotFoundError:
    record = 0

window = tk.Tk()
window.title("Кликер")
window.geometry("300x250")

score = 0

label_score = tk.Label(window, text=f"Очки: {score}", font=("Arial", 14))
label_score.pack(pady=10)

label_record = tk.Label(window, text=f"Рекорд: {record}", font=("Arial", 12), fg="blue")
label_record.pack(pady=5)

def update_score():
    global score, record
    score += 1
    label_score.config(text=f"Очки: {score}")
    
    if score > record:
        record = score
        label_record.config(text=f"Рекорд: {record}")
        save_record()

def save_record():
    with open("record.txt", "w") as file:
        file.write(str(record))

button = tk.Button(
    window, 
    text="Кликни меня!", 
    width=20, 
    height=2,
    command=update_score
)
button.pack(pady=10)

def key_event(event):
    if event.keysym in ["Return", "space"]:
        update_score()

window.bind("<Key>", key_event)

def reset_score():
    global score
    score = 0
    label_score.config(text=f"Очки: {score}")

reset_button = tk.Button(
    window,
    text="Сбросить счёт",
    width=20,
    height=1,
    command=reset_score,
)
reset_button.pack(pady=5)

def on_closing():
    save_record()
    window.destroy()

window.protocol("WM_DELETE_WINDOW", on_closing)

window.mainloop()
```

```python
import json
import os
import tkinter as tk
from tkinter import messagebox


def add_product():
    product = entry_product.get().strip()
    
    if not product:
        messagebox.showwarning("Предупреждение", "Введите название продукта!")
        return
    
    if product in listbox.get(0, tk.END):
        messagebox.showwarning("Предупреждение", f"Продукт '{product}' уже существует!")
        return
    
    listbox.insert(tk.END, product)
    entry_product.delete(0, tk.END)
    save_products()


def delete_product():
    try:
        selected_index = listbox.curselection()[0]
        listbox.delete(selected_index)
        save_products()
    except IndexError:
        messagebox.showwarning("Предупреждение", "Выберите продукт для удаления!")


def clear_all():
    if messagebox.askyesno("Подтверждение", "Вы уверены, что хотите очистить весь список?"):
        listbox.delete(0, tk.END)
        save_products()


def save_products():
    products = list(listbox.get(0, tk.END))
    
    try:
        with open("products.json", 'w', encoding='utf-8') as file:
            json.dump(products, file, ensure_ascii=False, indent=4)
    except Exception as e:
        messagebox.showerror("Ошибка", f"Не удалось сохранить файл:\n{str(e)}")


def load_products():
    if not os.path.exists("products.json"):
        return
    
    try:
        with open("products.json", 'r', encoding='utf-8') as file:
            products = json.load(file)
        
        listbox.delete(0, tk.END)
        
        for product in products:
            listbox.insert(tk.END, product)
            
    except json.JSONDecodeError:
        messagebox.showerror("Ошибка", "Файл повреждён или имеет неверный формат!")
    except Exception as e:
        messagebox.showerror("Ошибка", f"Не удалось загрузить файл:\n{str(e)}")


def on_enter_key(event):
    add_product()


root = tk.Tk()
root.title("Список продуктов")
root.geometry("400x500")
root.resizable(False, False)

top_frame = tk.Frame(root)
top_frame.pack(pady=10)

tk.Label(top_frame, text="Название продукта:", font=("Arial", 10)).pack(side=tk.LEFT, padx=5)
entry_product = tk.Entry(top_frame, width=25, font=("Arial", 10))
entry_product.pack(side=tk.LEFT, padx=5)
entry_product.bind("<Return>", on_enter_key)

btn_add = tk.Button(top_frame, text="Добавить", command=add_product, 
                    bg="#4CAF50", fg="white", font=("Arial", 10))
btn_add.pack(side=tk.LEFT, padx=5)

list_frame = tk.Frame(root)
list_frame.pack(pady=10, padx=20, fill=tk.BOTH, expand=True)

tk.Label(list_frame, text="Список продуктов:", font=("Arial", 12, "bold")).pack(anchor=tk.W)

scrollbar = tk.Scrollbar(list_frame)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

listbox = tk.Listbox(list_frame, yscrollcommand=scrollbar.set, 
                        font=("Arial", 10), height=15, selectmode=tk.SINGLE)
listbox.pack(fill=tk.BOTH, expand=True)
scrollbar.config(command=listbox.yview)

bottom_frame = tk.Frame(root)
bottom_frame.pack(pady=10)

btn_delete = tk.Button(bottom_frame, text="Удалить выделенный", 
                        command=delete_product, 
                        bg="#f44336", fg="white", font=("Arial", 10))
btn_delete.pack(side=tk.LEFT, padx=5)

btn_clear = tk.Button(bottom_frame, text="Очистить всё", 
                        command=clear_all, 
                        bg="#FF9800", fg="white", font=("Arial", 10))
btn_clear.pack(side=tk.LEFT, padx=5)

load_products()

root.mainloop()
```
