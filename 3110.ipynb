```python
#Шифр Цезарь
alphabet = 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩъыьЭЮЯ'

text = input("Введите слово для зашифрования: ").upper()
key = int(input("Введите ключ (число): "))

result = ""

for char in text:
    if char in alphabet:
        index = alphabet.index(char)
        new_index = (index + key) % len(alphabet)
        result += alphabet[new_index]
    else:
        result += char

print("Результат шифрования:", result)
```

```python
#Вижинер на русском
text = str(input("Введите слово для зашифровки"))
key = str(input("Введите ключ (слово): "))
result = ""

text = text.upper().replace(" ", "")
key = key.upper()

for i in range(len(text)):
    text_num = ord(text[i]) - ord('А')
    key_num = ord(key[i % len(key)]) - ord('А')
    new_num = (text_num + key_num) % 33
    result += chr(ord('А') + new_num)

print(f"Зашифровано: {result}")
```

```python
alphabet = 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'
alphabet_length = len(alphabet)

text = str(input("Введите слово для зашифрования: ")).upper()
key = str(input("Введите ключ (слово): ")).upper()

result = ''
key_index = 0

for char in text:
    if char in alphabet:
        text_index = alphabet.index(char)
        key_char = key[key_index % len(key)]
        key_shift = alphabet.index(key_char)
        new_index = (text_index + key_shift) % alphabet_length
        result += alphabet[new_index]
        key_index += 1
    else:
        result += char

print("Зашифрованный текст:", result)
```

```python
import math

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    try:
        return x / y
    except ZeroDivisionError:
        return "Ошибка: деление на ноль!"
    
def power(x, y):
    return x ** y

def square_root(x):
    if x >= 0:
        return math.sqrt(x)
    else:
        return "Ошибка: нельзя извлечь квадратный корень из отрицательного числа."

print("Выберите операцию:")
print("1. Сложение")
print("2. Вычитание")
print("3. Умножение")
print("4. Деление")
print("5. Возведение в степень")
print("6. Корень квадратный")

choice = input("Введите номер выбранной операции (1-6): ")

try:
    num1 = float(input("Введите первое число: "))
except ValueError:
    print("Ошибка: введите число.")

try:
    num2 = float(input("Введите второе число: "))
except ValueError:
    print("Ошибка: введите число.")

if choice == '1':
    print(f"{num1}+{num2}={add(num1,num2)}")

elif choice == '2':
    print(f"{num1}-{num2}={subtract(num1,num2)}")

elif choice == '3':
    print(f"{num1}*{num2}={multiply(num1,num2)}")

elif choice == '4':
    result = divide(num1, num2)
    if isinstance(result, str):
        print(result)
    else:
        print(f"{num1}/{num2}={result}")

elif choice == "5":
    print(f"{num1}**{num2}={power(num1, num2)}")

elif choice == "6":
    result1 = square_root(num1)
    result2 = square_root(num2)
    print(result1)    
    print(result2)

else:
    print("Некорректный ввод!")
```

```python
a = 10
b = 0


list = []

for i in range(a):
    try:
        x = int(input(f"list[{i}] = "))
        list.append(x)
    except ValueError:
        print("Ошибка: неправильно набранное число")
print("Список", list)

list_summa = sum(x for x in list if x > 0)
print("Сумма положительных чисел:", list_summa)

for x in list:
    if x % 2 == 0:
        b += 1
print("Кол-во чётныx элементы:", b)
    
max_index = 0
for i in range(1, len(list)):
    if list[i] > list[max_index]:
        max_index = i
print(f"Максимум: {list[max_index]}, индекс максимума: {max_index}")

print("Обратный список: ")
for i in range(len(list) - 1, -1, -1):
    print(list[i], end=' ')

list_new = [x for x in list if x >= 0]
print("Положительные элементы:", list_new)

y = [list[-1]] + list[:-1]
print("Последний элемент становится на первое место: ", y)

filtered = [x for x in list if x != 0]
alternates = True
for i in range(len(filtered) - 1):
    if (filtered[i] > 0 and filtered[i+1] > 0) or (filtered[i] < 0 and filtered[i+1] < 0):
        alternates = False
        break

print("YES" if alternates and len(filtered) > 1 else "NO")
```
